```python
def is_prime(num):
    sq = int(num ** 0.5)
    for i in range(2,sq + 1):
        if num % i == 0:
            return False
    return True
print(is_prime(25))
print(is_prime(71))
```

```python
def reverse_string(s): #learned slicing [::-1]
    if len(s) == 1:
        return s
    newStr = ""
    for i in range(1,len(s) + 1):
        newStr += s[-i]
    return newStr
print(reverse_string("emir"))
print(reverse_string("abcdefefdc"))
```

```python
def count_words(text):
    ourList = text.split()
    ourDict = {}
    for harf in ourList:
        ourDict[harf] = ourDict.get(harf,0) + 1
    return ourDict
print(count_words("emir emir abc asd emir asd a emir abc asd asd")) 
```

```python
def flatten_list(lst):
    flat = []
    for item in lst:
        if isinstance(item, list):
            flat.extend(flatten_list(item))
        else:
            flat.append(item)
    return flat

print(flatten_list([1, [2, 3], [4, [5, 6]]])) 
```

```python
def top_k_frequent(nums, k):
    freqDict = {}
    for harf in nums:
        freqDict[harf] = freqDict.get(harf,0) + 1
    #bucket sort
    buckets = [[] for _ in range(len(nums) + 1)]
    for n, c in freqDict.items():
        buckets[c].append(n)

    res = []
    for i in range(len(buckets) - 1, 0 , -1):
        for n in buckets[i]:
            res.append(n)
            if len(res) == k:
                return res
print(top_k_frequent([3,3,4,4,4,5,5,5,5,6,6,6,6,6], 2))
    
```

```python
def normalize_v(vector):
    norm = 0
    for i in vector:
        norm += i ** 2
    norm = norm ** 0.5
    if norm == 0:
        return vector
    return [ i / norm for i in vector]

print(normalize_v([3,4]))
print(normalize_v([5,12]))
print(normalize_v([7,24]))
print(normalize_v([0,0]))
```

```python
###sorted, with open() structure/modes (r,w,a) learned. 
with open("alice30.txt", "r")as file:
    text = file.read()
    text = text.lower()
    allWords = text.split()

    freqDict = {}
    for word in allWords:
        if word in freqDict:
            freqDict[word]+= 1
        else: 
            freqDict[word] = 1
    
    top10 = sorted(freqDict.items(), key=lambda x: x[1], reverse=True)[:10]
    for word, count in top10:
        print(word, count)
```
